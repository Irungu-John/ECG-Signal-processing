**Average time it takes for a Bus to complete a Route**

```python
df2['start'] = pd.to_datetime(df2[ 'TripStartTime'])   #Define start and stop between TripStartTime and TripEndTime 
df2['stop'] = pd.to_datetime(df2[ 'TripEndTime'])
```

```python
df2['delta'] = df2['stop'] - df2['start']  #Difference between start and stop 
```

```python
df2['delta']  #Bustrips time differnce
```

```python
df2['delta'].sum()  #The sum of the bus trips
```

```python
average_time = df2['delta'].sum() /7  #Average time it take for a Bus to complete a Route
print(average_time)          
```

**Average time** it takes for a Bus to complete a route is **46 minutes and 51 seconds**

**Map Plot Visualization of BlockNumbers and Locations**

```python
df2 = df2[["Lat", "Lon", "BlockNumber"]]  #Mapping the locations and the BlockNumber
```

```python
map = folium.Map(location=[df2.Lat.mean(), df2.Lon.mean()], zoom_start=11, control_scale=True) 
#Creating the Map  using latitude and longitude coordinates
```

```python
for index, location_info in df2.iterrows():      #Adding the Transit Route  Points
    folium.Marker([location_info["Lat"], location_info["Lon"]], popup=location_info["BlockNumber"]).add_to(map)
    
```
